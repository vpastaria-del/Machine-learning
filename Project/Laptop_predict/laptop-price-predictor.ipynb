```python
# 1. Convert Memory to string and remove decimal points
df["Memory"] = df["Memory"].astype(str).replace("\.0", "", regex=True)

# 2. Remove units
df["Memory"] = df["Memory"].str.replace("GB", "")
df["Memory"] = df["Memory"].str.replace("TB", "000")

# 3. Split into two parts (for dual drives)
new = df["Memory"].str.split("+", n=1, expand=True)

df["first"] = new[0].str.strip()
df["second"] = new[1].fillna("0").str.strip()

# helper to build flags & capacities
def split_drive(col):
    # flags for drive type
    hdd = df[col].str.contains("HDD", na=False).astype(int)
    ssd = df[col].str.contains("SSD", na=False).astype(int)
    hybrid = df[col].str.contains("Hybrid", na=False).astype(int)
    flash = df[col].str.contains("Flash", na=False).astype(int)

    # keep only digits for capacity
    cap = df[col].str.replace(r"\D", "", regex=True)
    cap = cap.replace("", "0").astype(int)

    return cap, hdd, ssd, hybrid, flash

# 47. Apply to first and second drive
df["first_cap"], df["Layer1HDD"], df["Layer1SSD"], df["Layer1Hybrid"], df["Layer1Flash_Storage"] = split_drive("first")
df["second_cap"], df["Layer2HDD"], df["Layer2SSD"], df["Layer2Hybrid"], df["Layer2Flash_Storage"] = split_drive("second")

# 8. Calculate total capacities
df["HDD"] = df["first_cap"] * df["Layer1HDD"] + df["second_cap"] * df["Layer2HDD"]
df["SSD"] = df["first_cap"] * df["Layer1SSD"] + df["second_cap"] * df["Layer2SSD"]
df["Hybrid"] = df["first_cap"] * df["Layer1Hybrid"] + df["second_cap"] * df["Layer2Hybrid"]
df["Flash_Storage"] = df["first_cap"] * df["Layer1Flash_Storage"] + df["second_cap"] * df["Layer2Flash_Storage"]

# 9. Drop helper columns
df.drop(
    columns=[
        "Memory", "first", "second",
        "first_cap", "second_cap",
        "Layer1HDD", "Layer1SSD", "Layer1Hybrid", "Layer1Flash_Storage",
        "Layer2HDD", "Layer2SSD", "Layer2Hybrid", "Layer2Flash_Storage",
    ],
    inplace=True,
)

print(df.head())
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np
import pickle

# 1. Correct Feature Lists
categorical_features = ['Company', 'TypeName', 'Cpu brand', 'Gpu brand', 'os']
numeric_features = ['Ram', 'Weight', 'Touchscreen', 'Ips', 'HDD', 'SSD']

# 2. Correct Preprocessor
step1 = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first'),
     categorical_features),
    ('num', StandardScaler(), numeric_features)
])

# 3. MLP Model
mlp = MLPRegressor(max_iter=600, random_state=42)

# 4. Pipeline
pipe = Pipeline([
    ('pre', step1),
    ('model', mlp)
])

# 5. Hyperparameter Search Space
params = {
    'model__hidden_layer_sizes': [(64, 32), (100, 50), (128, 64, 32)],
    'model__activation': ['relu', 'tanh'],
    'model__solver': ['adam'],
    'model__alpha': [0.0001, 0.001, 0.01],
    'model__learning_rate_init': [0.001, 0.01]
}

# 6. Training
print("Training Neural Network...")
search = RandomizedSearchCV(pipe, params, n_iter=10, cv=3, scoring='r2', n_jobs=-1, verbose=1)
search.fit(X_train, y_train)

# 7. Evaluate
best_model = search.best_estimator_
y_pred = best_model.predict(X_test)

print('R2 Score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))

# 8. Save Model
#pickle.dump(best_model, open('pipe.pkl', 'wb'))
#print("Saved new model as pipe.pkl")
```

```python
# ================================
# Deep MLP with Keras on log-prices
# ================================
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
import numpy as np

# 1. Preprocess features using SAME transformer as other models
#    Fit on training, transform both train and test
X_train_dl = pre.fit_transform(X_train)
X_test_dl = pre.transform(X_test)

input_dim = X_train_dl.shape[1]
print("Deep MLP input_dim:", input_dim)

# 2. Define the deep MLP architecture
model_dl = keras.Sequential([
    layers.Input(shape=(input_dim,)),

    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.4),

    layers.Dense(128, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    layers.Dense(64, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.2),

    # Output is log-price (same space as y_train / y_test)
    layers.Dense(1)
])

model_dl.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0005),
    loss='mse',
    metrics=[keras.metrics.MeanAbsoluteError(name="mae")]
)

early_stop = keras.callbacks.EarlyStopping(
    patience=25,
    restore_best_weights=True,
    monitor='val_loss'
)

history_dl = model_dl.fit(
    X_train_dl, y_train,          # <-- log-prices directly
    validation_split=0.2,
    epochs=300,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# 3. Evaluate in log-price space
y_pred_log = model_dl.predict(X_test_dl).ravel()

print("Deep MLP Test R2 (log-price):",
      r2_score(y_test, y_pred_log))
print("Deep MLP Test MAE (log-price):",
      mean_absolute_error(y_test, y_pred_log))
print("Deep MLP Test RMSE (log-price):",
      root_mean_squared_error(y_test, y_pred_log))

# 4. OPTIONAL: Convert back to actual prices for interpretation
y_test_price = np.exp(y_test)
y_pred_price = np.exp(y_pred_log)

print("Deep MLP Test R2 (log-price):",
      r2_score(y_test, y_pred_log))
dl_mlp_r2_score = r2_score(y_test, y_pred_log)
print("Deep MLP Test MAE (log-price):",
      mean_absolute_error(y_test, y_pred_log))
dl_mlp_mean_absolute_error = mean_absolute_error(y_test, y_pred_log)
print("Deep MLP Test RMSE (log-price):",
      root_mean_squared_error(y_test, y_pred_log))
dl_mlp_root_mean_squared_error = root_mean_squared_error(y_test, y_pred_log)
```
